```python
# Task 3:Text generation with markov chains
text = """
Once upon a time, in a magical forest, there lived a wise old owl.
The stars shone brightly over the enchanted mountains.
A dragon flew across the sky, leaving a trail of sparkling fire.
In the ancient library, books whispered secrets of the past.
The mysterious river glowed under the moonlight.
"""
```

```python
import random

def build_markov_chain(text, n=1):
    words = text.split()
    chain = {}

    for i in range(len(words) - n):
        key = tuple(words[i:i+n])
        next_word = words[i+n]
        if key not in chain:
            chain[key] = []
        chain[key].append(next_word)

    return chain


markov_chain = build_markov_chain(text, n=1)
```

```python
def generate_text(chain, length=50, n=1):

    start = random.choice(list(chain.keys()))
    output_words = list(start)

    for _ in range(length - n):
        key = tuple(output_words[-n:])
        if key in chain:
            next_word = random.choice(chain[key])
            output_words.append(next_word)
        else:
            break  # Stop if no next word

    return " ".join(output_words)

# Generate text
generated_text = generate_text(markov_chain, length=50, n=1)
print(generated_text)
```
